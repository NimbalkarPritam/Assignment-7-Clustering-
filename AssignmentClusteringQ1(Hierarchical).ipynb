# Hierarchical Clustering

```python
# import hierarchical clustering libraries
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import plotly.figure_factory as ff
```

```python
crime=pd.read_csv('E:\My folder\ExcelR(documents)\Assignments\Clustering-7\crime_data.csv')
```

```python
crime.head()
```

```python
crime.shape
```

```python
#Normalazation Function
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

# Dendrogram

```python
# Create Dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm,method='single'))
```

```python
# Create Dendrogram(Complete)
dendrogram = sch.dendrogram(sch.linkage(df_norm,method='complete'))
```

```python
# Create Dendrogram(Average)
dendrogram = sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
# Create Dendrogram(Centriod)
dendrogram = sch.dendrogram(sch.linkage(df_norm,method='ward'))
```

# Kmeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np
```

```python
crime=pd.read_csv('E:\My folder\ExcelR(documents)\Assignments\Clustering-7\crime_data.csv')
```

```python
crime.head()
```

```python
# Normalization Function
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(crime.iloc[:,1:])
```

```python
df_norm.head(10)
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Selecting 3 clusters
model=KMeans(n_clusters=3) 
model.fit(df_norm)

model.labels_ # getting the labels of clusters assigned to each row 
```

```python
md=pd.Series(model.labels_)  # converting numpy array into pandas series object 
crime['clust']=md # creating a  new column and assigning it to new column 
df_norm.head()
```

```python
crime.iloc[:,1:5].groupby(crime.clust).mean()
```

# DBSCAN 

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import .csv file and convert it to a DataFrame object
crime = pd.read_csv('E:\My folder\ExcelR(documents)\Assignments\Clustering-7\crime_data.csv');
crime.head()
```

```python
print(crime.info())
```

```python
crime1=pd.DataFrame(crime)
df=crime1[crime1.columns[1:5]]
df.head()
```

```python
array=df.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([df,cl],axis=1)
```
